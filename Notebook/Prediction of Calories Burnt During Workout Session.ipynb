```python
fig, axes = plt.subplots(4, 3, figsize=(12, 12))
axes = axes.flatten()
plt.suptitle('Comparative Analysis of Basic Features against Calories', fontsize=20, fontweight='bold')
for i, (category, title, top_n) in enumerate(categories):
    ax = axes[i]
    grouped_data = df.groupby(category)['Calories'].mean().sort_values(ascending=False)
    if top_n:
        grouped_data=grouped_data.head(top_n)
    grouped_data.plot.bar(ax = axes[i], color='orangered')
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.set_ylabel('Calories')
    ax.tick_params(axis='x', rotation=45)

for ax in range(len(categories), len(axes)):
    fig.delaxes(axes[ax])

plt.tight_layout()
plt.show()
```

### Feature Engineering

```python
categories = [
    ('User_ID', 'User_ID Vs Calories', 10),
    ('Gender', 'Gender Vs Calories', None),
    ('Age', 'Age Vs Calories', 10),
    ('Height', 'Height Vs Calories', 10),
    ('Weight', 'Weight Vs Calories', 10),
    ('Duration', 'Duration Vs Calories', 10),
    ('Heart_Rate', 'Heart_Rate Vs Calories', 10),
    ('Body_Temp', 'Body_Temp Vs Calories', 10)
]
```

```python
df['Age_Category'] = pd.cut(df["Age"],
                           bins=[0, 15, 18, 30, 45, 60, 100],
                            labels=["Teen", "Young Adult", "Full Adult", "Middle Age", "Old", "Aged"]
                           )

df["Height_Category"] = pd.cut(df["Height"],
                        bins=[0, 136, 155, 180, 200],
                         labels=["Short", "Medium", "Tall", "Very Tall"]
                        )

df["Weight_Category"] = pd.cut(df["Weight"],
                        bins=[0, 40, 52, 70, 115, 200],
                         labels=["Under Weight", "Low Weight", "Okay", "Normal Weight", "Obese"]
                        )

df["Heart_Rate_Range"] = pd.cut(df["Heart_Rate"],
                        bins=[0, 60, 100, 120, 150],
                         labels=["Low", "Normal", "High", "Tarchycardia"]
                        )

df["Body_Temperature_Category"] = pd.cut(df["Body_Temp"],
                        bins=[0, 35, 37, 45],
                         labels=["Low", "Normal", "High"]
                        )
```

## Data Splitting

With data splitting, I want to identify my dependent and independent variable. Y is my target variable which is my dependent variable. My independent variable is x. In this dataset my dependent variable is going to be my Calories which is affected by people's age, Height, Weight and all other factors. These other factors are my independent variable

```python
df.head()
```

```python
x = df.drop(columns=['Calories', 'log_Calories']) #my independent variable
y= df['log_Calories'] #my dependent variable
```

```python
x
```

```python
y
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_inputs, test_input, train_target, test_target, = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Data Preprocessing

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
train_inputs
```

```python
Num_col = df.select_dtypes(include=np.number).columns
cat_c
```

```python
cat_col = ['Gender', 'Age_Category', 'Height_Category', 'Weight_Category', 'Heart_Rate_Range', 'Body_Temperature_Category']
```

```python
encoder = OneHotEncoder(drop="first")
train_cat = encoder.fit_transform(train_inputs[cat_col])
test_cat = encoder.transform(test_input[cat_col])
```

```python
train_cat
test_cat.toarray()
```

### Scaling

```python
scaler = StandardScaler()
```

```python
Num_col = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']
```

```python
train_num = scaler.fit_transform(train_inputs[Num_col])
test_num = scaler.transform(test_input[Num_col])
```

```python
train_processed = np.hstack((train_num, train_cat.toarray()))
test_processed = np.hstack((test_num, test_cat.toarray()))
test_processed
train_processed
```

### Data Modelling

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(train_processed, train_target)
model.fit(test_processed, test_target)
```

### Model Prediction

```python
train_pred = model.predict(train_processed)
test_pred = model.predict(test_processed)
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train_mse = mean_squared_error(train_target, train_pred)
test_mse = mean_squared_error(test_target, test_pred)
train_mse, test_mse
```

# Model Performance

Here I'll use regression metrics and tools such as the Mean Absolute error, Mean Squared Error, Root Mean Squared error to evaluate the performance of my model

```python
train_score = r2_score(train_target, train_pred)
test_score = r2_score(test_target, test_pred)
```

```python
train_score, test_score
```

#### 
**Training RÂ² Score (0.956):**

This means that approximately 95.6% of the variance in the training data's target variable (log_Calories) is explained by the model. This indicates a good fit to the training data, suggesting that the model captures a significant amount of the underlying patterns.

**Test RÂ² Score (0.958):**

This means that approximately 95.8% of the variance in the test data's target variable is explained by the model. This is slightly better than the training RÂ² score, which generally indicates that the model generalizes well to new, unseen data. It suggests that the model performs slightly better on the test set compared to the training set.
